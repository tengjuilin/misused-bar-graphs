## Original linear data (+/+)

```python
# data: original linear
# axis: linear
# ymin: zero
x1 = 1
x2 = 1.5
category = ['A', 'B']
fig, axs = plt.subplots(1, 2, figsize=(4, 4))
axs[0].bar(category, [x1, x2])
bars = axs[1].bar(category, [x1, x2])
axs[0].set_title('True')
axs[1].set_title('Visualized')
plt.tight_layout()

ymin, _ = axs[1].get_ylim()
fig.canvas.draw()
r = fig.canvas.get_renderer()
y1, y2 = [bar.get_window_extent(r).height for bar in bars]
lie_factor = (y2 - y1) / (x2 - x1) * x1 / y1
fig.supxlabel(f'Lie Factor = {round(lie_factor, 2)}', fontsize=16, fontweight='bold');
```
