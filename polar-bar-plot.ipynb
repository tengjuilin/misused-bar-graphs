## Circular Bar Chart

```python
size = 100
np.random.seed(0)
df = pd.DataFrame(
        {
            'Name': ['item ' + str(i) for i in list(range(1, size+1)) ],
            'Value': np.arange(0, size),
        })

# Reorder the dataframe
df = df.sort_values(by=['Value'])
```

```python
base = 50
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})
for i in range(df.shape[0]):
    theta = 2 * np.pi / (df.shape[0] + 1) * i
    height = base + i
    ax.scatter([theta], [height], s=10, color='k')

for i in [25, 50]:
    theta = 2 * np.pi / (df.shape[0] + 1) * i
    height = base + i
    ax.scatter([theta], [height], color='r')

# ax.axis('off')
```

```python
base = 50
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})
for i in range(df.shape[0]):
    theta = 2 * np.pi / (df.shape[0] + 1) * i
    height = base + i
    ax.plot([theta, theta], [base, height], 'k-', lw=1.5, zorder=0)
    ax.scatter([theta], [height], s=10, color='k', zorder=2)

for i in [25, 50, 75, 100]:
    theta = 2 * np.pi / (df.shape[0] + 1) * i
    height = base + i
    ax.plot([theta, theta], [base, height], 'r-', lw=1.5, zorder=0)
    ax.scatter([theta], [height], s=10, color='r', zorder=2)

ax.spines['polar'].set_visible(False)
ax.set_xticks([])
yticklabels = np.array([0, 25, 50, 75, 100])
yticks = yticklabels + 50
ax.set_yticks(yticks)
ax.set_yticklabels(yticklabels)
ax.xaxis.grid(False)
ax.yaxis.grid(True, color='k', linestyle='-', lw=0.4, alpha=0.2, zorder=1)
ax.yaxis.set_tick_params(labelsize=10)
ax.set_rlabel_position(-7)
# ax.axis('off')
# plt.savefig('a.svg', format='svg', transparent=True)
```

## Radial Bar Plot

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

############# subplot 1 #############
data = np.arange(10, 80, 10)
n = len(data)
colors = [cm.plasma_r(i / n) for i in range(n)]
axs[0].bar(
    np.arange(7), 
    np.arange(10, 80, 10),
    color=colors,
)
axs[0].set_xticklabels(['', 'A', 'B', 'C', 'D', 'E', 'F', 'G'])
axs[0].set_ylim(0, 80)
axs[0].set_ylabel('Numerical value')

############# subplot 2 #############
labels = [
    'A = 10', 
    'B = 20', 
    'C = 30', 
    'D = 40', 
    'E = 50', 
    'F = 60', 
    'G = 70',
]
data = np.arange(10, 80, 10)
# number of data points
n = len(data)
# find max value for full ring
k = 10 ** int(np.log10(max(data)))
m = k * (1 + max(data) // k) 
# radius of donut chart
r = 1.5
# calculate width of each ring
w = r / n 

# create colors along a chosen colormap
colors = [cm.plasma_r(i / n) for i in range(n)]

# create rings of donut chart
for i in range(n):
    innerring, _ = axs[1].pie(
        [m - data[i], data[i]], 
        radius = r - i * w, 
        startangle = 90, 
        labels = ["", labels[i]], 
        labeldistance = 1 - 1 / (1.5 * (n - i)), 
        textprops = {"alpha": 0}, 
        colors = ["white", colors[i]],
    )
    plt.setp(
        innerring, 
        width = w, 
        edgecolor = "white",
    )

axs[1].axis("equal")
axs[1].set_box_aspect(1)
axs[1].legend(bbox_to_anchor=(0, 1), loc='upper left')
axs[1].set_title('Numerical value in radial bar graph\n')

############# subplot 3 #############
labels = list("ABCDEFG")[::-1]
data = np.arange(10, 80, 10)[::-1]
# number of data points
n = len(data)
# find max value for full ring
k = 10 ** int(np.log10(max(data)))
m = k * (1 + max(data) // k) 
# radius of donut chart
r = 1.5
# calculate width of each ring
w = r / n 

# create colors along a chosen colormap
colors = [cm.plasma_r(i / n) for i in range(n)][::-1]

# create rings of donut chart
for i in range(n):
    innerring, _ = axs[2].pie(
        [m - data[i], data[i]], 
        radius = r - i * w, 
        startangle = 90, 
        labels = ["", labels[i]], 
        labeldistance = 1 - 1 / (1.5 * (n - i)), 
        textprops = {"alpha": 0}, 
        colors = ["white", colors[i]],
    )
    plt.setp(
        innerring, 
        width = w, 
        edgecolor = "white",
    )

axs[2].axis("equal")
axs[2].set_title('Numerical value in radial bar graph\n')

for i in range(3):
    axs[i].set_box_aspect(1)

lw = 1.5
ls = '--'
linecolor = 'w'
for i in range(1, 3):
    axs[i].plot([0, 0], [0, r], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, r], [0, 0], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, 0], [0, -r], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, -r], [0, 0], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, np.sqrt(r)], [0, np.sqrt(r)], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, np.sqrt(r)], [0, -np.sqrt(r)], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, -np.sqrt(r)], [0, np.sqrt(r)], color=linecolor, lw=lw, ls=ls)
    axs[i].plot([0, -np.sqrt(r)], [0, -np.sqrt(r)], color=linecolor, lw=lw, ls=ls)
    axs[i].text(0.1, r + 0.2, '0')
    axs[i].text(r + 0.1, 0, '20')
    axs[i].text(0, -(r + 0.2), '40')
    axs[i].text(-(r + 0.2), 0, '60')
    axs[i].text(-0.2, r + 0.2, '80')
    axs[i].text(np.sqrt(r), np.sqrt(r), '10')
    axs[i].text(np.sqrt(r), -np.sqrt(r), '30')
    axs[i].text(-np.sqrt(r), -np.sqrt(r), '50')
    axs[i].text(-np.sqrt(r), np.sqrt(r), '70')

plt.tight_layout(pad=2)
fig.savefig('figures/subpanels/ex-fig-10-radial-bar-graphs.pdf')
```
